```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class GBM:
    def __init__(self, S0, mu, sigma, T, n):
        self.mu = mu
        self.sigma = sigma
        self.n = n
        self.S0 = S0
        self.T = T 
    
    def sde(self):
        
        dt = self.T/self.n
        t = np.linspace(0,self.T,self.n + 1)
        W = np.cumsum(np.random.normal(0, np.sqrt(dt), self.n + 1))
        x = (self.mu - 0.5*self.sigma**2)*t + self.sigma*W
        S = self.S0 * np.exp(x)
        
        return t, S 
        
```

```python
g = GBM(S0 = 10, T = 2, n = 1000, mu = 0.1, sigma = 0.05)
```

```python
time,stock_price = g.sde()
```

```python
plt.plot(time,stock_price)
plt.xlabel('Time (t)')
plt.ylabel('Stock Price S(t)')
plt.title('Geometric Brownian Motion')
plt.show()
```
