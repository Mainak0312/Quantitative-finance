## Equity ACDC Product: Reframed Description and Model Framework

**Product Overview**

Equity ACDC is an over-the-counter (OTC) structured product, offered by an issuer to investors, which obliges buyers to buy or sell shares of a specified underlying security at a predetermined strike price. Settlement is periodic, typically monthly, over the contract's life (for example, 12 periods for a one-year expiry)[1][2].

**Key Notation and Definitions**

- $$ t_i $$: The $$i$$th settlement date, with $$N$$ total periods (e.g., $$N = 12$$ for monthly settlement over a year).
- $$ n_i $$: Number of days in the $$i$$th period.
- $$ t_{i,j} $$: The $$j$$th accumulation date in the $$i$$th period.
- $$ m $$: Number of shares per day.
- $$ K $$: Strike price.
- $$ S_i $$: Stock price on settlement date $$t_i$$.
- $$ B $$: Knock-out/Autocall barrier.
- $$ G $$: Gear (multiplier for shares when conditions are not met).
- $$ \delta $$: Indicator for accumulator ($$1$$) or decumulator ($$-1$$).

**Payoff Structure**

The payoff at expiry is defined as:
$$
p_{\text{payoff}} = \sum_{i=1}^N X_i A_i
$$
where:
- $$ X_i $$: Number of shares accumulated in the $$i$$th period, given by:
  $$
  X_i = \sum_{j=1}^{n_i} \left[ m \cdot \mathbf{1}\{\delta(s_{i,j} - K) \geq 0\} + G \cdot m \cdot \mathbf{1}\{\delta(s_{i,j} - K) < 0\} \right] \cdot \mathbf{1}(s_{i,j} < B)
  $$
- $$ A_i = \delta(S_i - K) $$
- $$ \delta = 1 $$ for accumulator, $$ -1 $$ for decumulator.

**Barrier Features**

- **Knock-out**: If the barrier is breached, only the days where this occurs are removed from the accumulation/decumulation schedule; the trade otherwise continues.
- **Autocall**: The trade terminates immediately upon the barrier being hit.
- **No-call period**: Barriers are not checked before a minimum accumulation period; shares accumulate regardless of barrier breaches during this time[1][2].

**Valuation Model Framework**

The present value (PV) of the ACDC contract at valuation date $$t$$ is the expected discounted value of the payoff:
$$
PV(t) = \mathbb{E} \left( \sum_{i=1}^{N} X_i A_i \cdot df(t_i, t) \right)
$$
where $$df(t_i, t)$$ is the discount factor from settlement date $$t_i$$ to valuation date $$t$$[1][2].

**Pricing Methods**

- **Monte Carlo Simulation**: Simulates future stock price paths (assuming a lognormal process with local or stochastic local volatility), tracks share accrual, and computes the equity leg at each period's settlement date.
- **PDE (Partial Differential Equation) Method**: Solves backward on a grid, computing PV for each spot/time combination. Adjustments are made for the probability of autocall/knock-out, using different approaches:
  - **None**: Ignores early autocall impact (least accurate).
  - **Accrual**: Accrues shares at current spot for the current period, period settlement for future periods (fastest, less accurate for large dividends).
  - **ProbStates**: Tracks the probability of autocall for each remaining date, adjusting accrual accordingly; converges to the theoretically correct price.
  - **ProbApprox**: Approximates autocall probabilities using analytic formulas for continuous barriers.
  - **ProbApproxFull**: Improves on ProbApprox by computing autocall probabilities for each intermediate date (more accurate, slower)[1][2].

```python
Gearing applies when the spot price is strictly below the strike in the case of an Accumulator, or strictly above the strike for a Decumulator. Shares accrue daily throughout the accumulation schedule and are settled at the Period End Date, with payment made on the corresponding Settlement Date. If an autocall event occurs on any accumulation date, the accrued shares are settled early on the Autocall Payment Date. Each date in the accumulation schedule specifies a barrier type, which is typically Autocall. However, it can also be a Knock-Out, in which case the trade continues even if the barrier is breached—though any day on which the barrier is crossed is excluded from share accrual. When a minimum accumulation period is configured, shares will continue to accumulate regardless of whether the barrier is breached, until that minimum tenor is completed.
```
