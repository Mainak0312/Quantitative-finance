```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as nrp
```

```python
def weiner_process(dt, n =1000): # dt is the variance and n is the number of samples
    
    # W(t=0)=0
    w = np.zeros(n+1) # initialize W(t) with zeros
    
    # we create N+1 timesteps: t=0,1,2,3...N
    t = np.linspace(0,n,n+1) 
    
    # we have to use cumulative sum: on every step the additional value is
    # drawn from a normal distribution with mean 0 and variance dt ... N(0,dt)
    # btw: N(0,dt) = sqrt(dt)*N(0,1) is used here!!!

    w[1:n+1] = np.cumsum(nrp.normal(0, np.sqrt(dt), n))
    return t, w
```

```python
def plot_data(time, data):
    plt.plot(time, data)
    plt.xlabel('time')
    plt.ylabel('weiner process')
    plt.title('weiner process')
    plt.show()
```

```python
time, data = weiner_process(0.1)
```

```python
plot_data(time, data)
```
