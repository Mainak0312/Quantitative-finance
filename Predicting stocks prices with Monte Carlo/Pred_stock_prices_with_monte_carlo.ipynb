```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# no of realizations
no_of_simulations = 1000
```

```python
def stock_monte_carlo(S0,mu, sigma, N = 251):
    result = []
    
    for _ in range(no_of_simulations):
        price = [S0]
        
        for _ in range(N):
            stock_price = price[-1] * np.exp((mu - 0.5*sigma**2) + sigma * np.random.normal())
            price.append(stock_price)
        result.append(price)  
    
    sim_data = pd.DataFrame(result)
    #  the given columns will contain the time series for a given realization
    sim_data = sim_data.T
    
    return sim_data
```

```python
sim_data = stock_monte_carlo(50, 0.0002, 0.01)
sim_data['mean'] = sim_data.mean(axis = 1)
```

```python
plt.plot(sim_data)
plt.show()
```

```python
plt.plot(sim_data['mean'])
plt.show()
```

```python
print("Stock price at the end of the year is $ %2f" % sim_data['mean'].tail(1))
```
