```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class weiner:
    def __init__(self, dt, mu, n):
        self.dt = dt
        self.mu = mu
        self.n = n
    
    def weiner_process(self):
        w = np.zeros(self.n+1)
        
        # time step
        t = np.linspace(0,self.n, self.n+1)
        
        w[1:self.n + 1] = np.cumsum(np.random.normal(self.mu, np.sqrt(self.dt), self.n))
        return t, w
    
   
```

```python
weiner_chck = weiner(0.1, 0, 1000)
```

```python
t,w = weiner_chck.weiner_process()
```

```python
plt.plot(t,w)
plt.show()
```
