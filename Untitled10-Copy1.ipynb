Euler-Maruyama Scheme:

$$S_{t+\delta t}\sim S_t(1+r\delta t+\sigma\phi\sqrt{\delta t})$$

```python
#Euler-Maruyama Scheme:
def simulate_path(s0, mu, sigma, horizon, timesteps, n_sims, seed=None):
    if seed is not None:
        np.random.seed(seed)
    dt = horizon / timesteps
    S = np.zeros((timesteps, n_sims))
    S[0, :] = s0
    for i in range(1, timesteps):
        w = np.random.standard_normal(n_sims)
        S[i, :] = S[i-1, :] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * w)
    return S
```

Milstein Scheme:

$$S_{t+\delta t}\sim S_t\left(1+r\delta t+\sigma\phi\sqrt{\delta t}+\dfrac{1}{2}\sigma^{2}\left(\phi^2-1\right)\delta t\right)$$


```python
#Milstein Scheme:
def simulate_path_milstein(s0, mu, sigma, horizon, timesteps, n_sims, seed=None):
    if seed is not None:
        np.random.seed(seed)
    
    dt = horizon / timesteps  # Time step
    S = np.zeros((timesteps, n_sims))  # Initialize stock price matrix
    S[0, :] = s0  # Set initial stock price
    
    for i in range(1, timesteps):
        w = np.random.standard_normal(n_sims)  # Standard normal random variable
        dW = np.sqrt(dt) * w  # Brownian motion increment
        
        # Milstein correction term
        correction = 0.5 * sigma**2 * S[i-1, :] * (dW**2 - dt)
        
        # Apply the Milstein update
        S[i, :] = S[i-1, :] + mu * S[i-1, :] * dt + sigma * S[i-1, :] * dW + correction
    
    return S
```

Simulate the value of an ACDC (Accumulator/Decumulator) contract using Monte Carlo.

In each period, we compute

$$
X_i = \sum_{j=1}^{n_i} \left[ m \cdot \mathbf{1}\{\delta(s_{i,j} - K) \geq 0\} + G \cdot m \cdot \mathbf{1}\{\delta(s_{i,j} - K) < 0\} \right] \cdot \mathbf{1}(s_{i,j} < B)
$$

and

$$
A_i = \delta (S_i - K)
$$

Then the total present value is given by

$$
PV(t) = \mathbb{E} \left( \sum_{i=1}^{N} X_i A_i \cdot df(t_i, t) \right)
$$

where the discount factor at the settlement date of period \(i\) is defined as

$$
df(t_i, t) = e^{-r t_i}.
$$



```python
def simulate_acdc(S0, r, sigma, T, N_paths, N_periods, days_per_year, m, K, B, G, contract='accumulator', seed=None):
    """
    Simulate the value of an ACDC (Accumulator/Decumulator) contract using Monte Carlo.
    
    For each period, the simulation:
      - Accumulates shares daily based on:
          if delta*(price - B) < 0:
              if delta*(price - K) >= 0: accumulate m shares
              else: accumulate G * m shares
      - At the period's settlement date, calculates the payoff using:
          A_i = delta * (S_settlement - K)
          Payoff_i = (accumulated shares in period) * A_i, discounted to present.
    
    The total payoff is the sum over all periods.
    
    Returns:
        price: float, the estimated price of the ACDC contract.
    """
    n_steps = int(days_per_year * T)  # Total simulation steps; for 1 year, n_steps = 252
    S = simulate_path(S0, r, sigma, T, n_steps, N_paths, seed)
    delta = 1 if contract.lower() == 'accumulator' else -1
    period_indices = np.linspace(0, n_steps, N_periods + 1, dtype=int)
    period_payoffs = np.zeros((N_paths, N_periods))
    
    for i in range(N_periods):
        start_index = period_indices[i]
        # For the final period, use the full period so that t_settlement = n_steps/days_per_year
        if i == N_periods - 1:
            settlement_index = n_steps - 1  # Last index of S (i.e., day 251, which represents 252 days)
            t_settlement = n_steps / days_per_year  # t_settlement = 252/252 = 1 year
            S_settlement = S[settlement_index, :]
        else:
            settlement_index = period_indices[i + 1] - 1
            t_settlement = (settlement_index + 1) / days_per_year  # +1 to count days correctly
            S_settlement = S[settlement_index, :]
        
        X_i = np.zeros(N_paths)
        days_in_period = settlement_index - start_index + 1
        
        for day in range(days_in_period):
            for sim in range(N_paths):
                price = S[start_index + day, sim]
                if delta * (price - B) < 0:
                    if delta * (price - K) >= 0:
                        X_i[sim] += m  
                    else:
                        X_i[sim] += G * m  
                else:
                    X_i[sim] += 0
        
        A_i = delta * (S_settlement - K)
        period_payoff = X_i * A_i
        discount_factor = np.exp(-r * t_settlement)
        period_payoffs[:, i] = period_payoff * discount_factor
        
        # Print discount factor and the average discounted payoff for this period
        avg_payoff = np.mean(period_payoffs[:, i])
        #print(f"Period {i+1}: t_settlement = {t_settlement*days_per_year:.3f} days, discount_factor = {discount_factor:.9f}, avg discounted payoff = {avg_payoff:.4f}")
    
    total_payoffs = np.sum(period_payoffs, axis=1)
    price = np.mean(total_payoffs)
    return price
```

```python
import numpy as np

def simulate_path(s0, mu, sigma_initial, smile, kappa, skew, horizon, timesteps, n_sims, seed=None):
    """
    Simulate stock price paths using a Stochastic Local Volatility model.
    
    Parameters:
        s0: Initial stock price.
        mu: Drift rate (annualized).
        sigma_initial: Initial volatility (annualized).
        smile: Smile parameter for local volatility adjustment.
        kappa: Mean reversion rate of the variance process.
        skew: Correlation between stock and variance processes.
        horizon: Time horizon (years).
        timesteps: Number of time steps.
        n_sims: Number of simulations.
        seed: Random seed for reproducibility.
    """
    if seed is not None:
        np.random.seed(seed)
    dt = horizon / timesteps
    S = np.zeros((timesteps, n_sims))
    v = np.zeros_like(S)
    # Initial variance (square of initial volatility)
    v0 = sigma_initial ** 2
    S[0, :] = s0
    v[0, :] = v0
    # Heston model parameters (assumed or derived)
    theta = v0  # Long-term variance (set to initial variance)
    xi = 0.1    # Volatility of volatility (assumed)
    
    for i in range(1, timesteps):
        # Generate correlated Brownian motions
        corr_matrix = [[1.0, skew], [skew, 1.0]]
        Z = np.random.multivariate_normal(mean=[0, 0], cov=corr_matrix, size=n_sims)
        Z1 = Z[:, 0]  # For stock process
        Z2 = Z[:, 1]  # For variance process
        
        # Update variance process
        v_prev = v[i-1, :]
        dv = kappa * (theta - v_prev) * dt + xi * np.sqrt(v_prev * dt) * Z2
        v_new = v_prev + dv
        v_new = np.maximum(v_new, 0)  # Ensure non-negative variance
        v[i, :] = v_new
        
        # Local volatility adjustment
        S_prev = S[i-1, :]
        L = 1 + smile * (S_prev / s0 - 1)
        sigma_t = L * np.sqrt(v_prev)
        
        # Update stock price
        drift = (mu - 0.5 * sigma_t**2) * dt
        diffusion = sigma_t * np.sqrt(dt) * Z1
        S[i, :] = S_prev * np.exp(drift + diffusion)
    
    return S

def simulate_acdc(S0, r, sigma_initial, smile, kappa, skew, T, N_paths, N_periods, days_per_year, m, K, B, G, contract='accumulator', seed=None):
    """
    Simulate the value of an ACDC contract using Stochastic Local Volatility.
    """
    n_steps = int(days_per_year * T)
    S = simulate_path(S0, r, sigma_initial, smile, kappa, skew, T, n_steps, N_paths, seed)
    delta = 1 if contract.lower() == 'accumulator' else -1
    period_indices = np.linspace(0, n_steps, N_periods + 1, dtype=int)
    period_payoffs = np.zeros((N_paths, N_periods))
    
    for i in range(N_periods):
        start_index = period_indices[i]
        if i == N_periods - 1:
            settlement_index = n_steps - 1
            t_settlement = n_steps / days_per_year
            S_settlement = S[settlement_index, :]
        else:
            settlement_index = period_indices[i + 1] - 1
            t_settlement = (settlement_index + 1) / days_per_year
            S_settlement = S[settlement_index, :]
        
        X_i = np.zeros(N_paths)
        days_in_period = settlement_index - start_index + 1
        
        for day in range(days_in_period):
            for sim in range(N_paths):
                price = S[start_index + day, sim]
                if delta * (price - B) < 0:
                    if delta * (price - K) >= 0:
                        X_i[sim] += m
                    else:
                        X_i[sim] += G * m
        
        A_i = delta * (S_settlement - K)
        period_payoff = X_i * A_i
        discount_factor = np.exp(-r * t_settlement)
        period_payoffs[:, i] = period_payoff * discount_factor
        
        avg_payoff = np.mean(period_payoffs[:, i])
        #print(f"Period {i+1}: t_settlement = {t_settlement*days_per_year:.3f} days, discount_factor = {discount_factor:.9f}, avg discounted payoff = {avg_payoff:.4f}")
    
    total_payoffs = np.sum(period_payoffs, axis=1)
    price = np.mean(total_payoffs)
    return price
```

The model prices ACDC trades by computing the expected discounted payoff accrued through daily share accumulations over a series of settlement periods. In each period, shares are accumulated or decumulated based on whether the underlying stocks price relative to a fixed strike crosses a thresholdwith the basic accumulation governed by a multiplier, and further adjusted by a gear factor if the movement is adverse. The structure accommodates knock-out and autocall provisions: if the barrier is breached, the corresponding days are either removed from the accumulation schedule or cause an early termination of the trade, while trades with no call period continue accumulating until a minimum period end date regardless of barrier hits. Valuation is then achieved using either Monte Carlo simulations, where evolving spot prices directly affect share accumulation, or by solving a backward partial differential equation that integrates conditional forward prices and autocall probability adjustments.
