```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def geo_ran_walk(s0, T, n, mu, sigma):
    dt = T/n
    t = np.linspace(0,T,n)
    w = np.random.standard_normal(n)
    w = np.cumsum(w) *np.sqrt(dt) # N(0,dt) = sqrt(dt) * N(0,1)
    x = np.exp((mu-0.5*sigma*sigma)*t + sigma*w)
    s = s0*x
    return t, s
```

```python
def plot_simulation(t, S):
    plt.plot(t, S)
    plt.xlabel('Time (t)')
    plt.ylabel('Stock Price S(t)')
    plt.title('Geometric Brownian Motion')
    plt.show()
```

```python
time, data = geo_ran_walk(10, 2, 1000, 0.1, 0.05)
```

```python
plot_simulation(time, data)
```
